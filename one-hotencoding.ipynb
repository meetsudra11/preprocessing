pandas generate output based on the data provided at that moment, it doesnot maintain the column ordering thus scikit learns one-hot encoding is preferred

```python
# Lets now see one-hot encoding using scikit learn library

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop('selling_price',axis=1),df['selling_price'],test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_train.head())
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import OneHotEncoder

Ohe = OneHotEncoder(dtype=np.int32) # matrix me values integer milegi 
# To avoid multicollinearity u use drop attribute in the onehotencoder class
# drop = 'first'

# Ohe = OneHotEncoder(drop='first',sparse=False)
```

```python
Ohe.fit_transform(X_train[['fuel','owner']])
```

```python
X_train_cat = Ohe.fit_transform(X_train[['fuel','owner']]).toarray()
X_train_cat

# if u use sparse parameter in the onehotencoder class u then dont need to convert into array then
```
